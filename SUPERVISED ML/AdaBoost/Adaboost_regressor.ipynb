```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('car_price_prediction_.csv')
df.head()
```

```python
df.info()
```

```python
df=df.drop('Car ID', axis=1)
```

```python
df.to_csv('cleaned_regression.csv')
```

```python
x=df.drop('Price', axis=1)
y=df['Price']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['Model']=le.fit_transform(x['Model'])
```

```python
num_feature=x.select_dtypes(exclude='str').columns 
cat_feature=x.select_dtypes(include='str').columns 
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
numeric_transformer=StandardScaler()
cat_transformer=OneHotEncoder()
preprocessor=ColumnTransformer(
    [
        ('OneHotEncoder', cat_transformer, cat_feature),
        ('StandardScaler', numeric_transformer, num_feature)
    ], remainder='passthrough'
)
x=preprocessor.fit_transform(x)
x=pd.DataFrame(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
models={
    "AdaBoostRegressor": AdaBoostRegressor(),
    "RandomForestRegressor":RandomForestRegressor(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "KNeighborsRegressor":KNeighborsRegressor(),
    "LinearRegression": LinearRegression()
}
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train, y_train)
```

```python
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)
```

```python
print(r2_score(y_train,y_train_pred))
print(mean_absolute_error(y_train, y_train_pred))
print(mean_squared_error(y_train, y_train_pred))
print(np.sqrt(mean_absolute_error(y_train,y_train_pred)))
```

```python
print(r2_score(y_test,y_test_pred))
print(mean_absolute_error(y_test, y_test_pred))
print(mean_squared_error(y_test, y_test_pred))
print(np.sqrt(mean_absolute_error(y_test,y_test_pred)))
```

```python
rf_param={
    "max_depth":[2,8,5,None,20],
    "max_features":[5,7,'auto',9],
    "min_samples_split":[2,8,15,20],
    "n_estimators":[100,200,500,1000]
}
adaboost_param={
    "n_estimators":[50,60,70,80,90],
    "learning_rate": [0.01, 0.1, 0.5, 1.0]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomcv=[
        ("RF", RandomForestRegressor(), rf_param),
        ("AB", AdaBoostRegressor(), adaboost_param)
    ]
```

```python
model_param={}
for name, model, params in randomcv:
    random=RandomizedSearchCV(estimator=model,
                              param_distributions=params,
                              cv=3,
                              n_iter=100,
                              verbose=2,
                              n_jobs=-1)
    random.fit(x_train,y_train)
```

```python
y_train_pred=random.predict(x_train)
y_test_pred=random.predict(x_test)
```

```python
print(r2_score(y_train,y_train_pred))
print(mean_absolute_error(y_train, y_train_pred))
print(mean_squared_error(y_train, y_train_pred))
print(np.sqrt(mean_absolute_error(y_train,y_train_pred)))
```

```python
print(r2_score(y_test, y_test_pred))
print(mean_absolute_error(y_test, y_test_pred))
print(mean_squared_error(y_test, y_test_pred))
```
