```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000, n_features=5, n_classes=2, random_state=42)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(x_train, y_train)
```

```python
y_pred=logistic.predict(x_test)
print(y_pred)
```

```python
logistic.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
score=accuracy_score(y_pred, y_test)
cm=confusion_matrix(y_pred, y_test)
print(classification_report(y_pred, y_test))
print(score)
print(cm)
```

HYPERPARAMETER TUNING

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_vlaue=[100,10,1.0,0.1,0.01]
solver=['newton_cg', 'lbfgs', 'liblinear','sag', 'saga']
```

```python
params=dict(penalty=penalty, C=c_vlaue, solver=solver)
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
grid.fit(x_train, y_train)
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
print(accuracy_score(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
random=RandomizedSearchCV(estimator=model, cv=5, param_distributions=params, scoring='accuracy')
```

```python
random.fit(x_train, y_train)
```

```python
y_pred=random.predict(x_test)
```

```python
random.best_params_
```

```python
print(accuracy_score(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```
