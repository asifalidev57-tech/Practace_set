```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('data_for_eda.xlsx')
df.head()
```

```python
df.info()
```

```python
df.head(2)
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1, inplace=True )
```

```python
df.head(1)
```

```python
df['Duration_Hour']=(df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int))
df['Duration_Min']=(df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int))
df.info()
```

```python
df.drop(('Duration'),axis=1, inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.head(2)
```

```python
df['Dep_Time'].astype=(int)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minn']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
df.head(20)
```

```python
df['Arrival_Minn'].apply(lambda x: x.split(' ')[0])
```

```python
df.head(2)
```

```python
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_Min']=df['Dep_Min'].astype(int)
```

```python
df.head(2)
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stops':1, '2 stops':2, '3 stops': 3, '4 stops':4, np.nan: 1})
```

```python
df.head(2)
```
