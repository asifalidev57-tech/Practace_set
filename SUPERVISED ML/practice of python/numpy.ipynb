```python
import pandas as pd 
data={
    "name":["ali","khan","asif"],
    "age":[12, 20, 19],
    "city":["mardan", "ibd", "mhmnd"]
}
df=pd.DataFrame(data)
df.iat[1, 0]

df['salary']=[9000,877,67790]
df['age']=df['age']+20
df['city']=df['city']+"  in pakistan"
df.drop('salary', axis=1, inplace=True)
print(df)
```

```python
import pandas as pd 
data={
    "name":["ali","khan","asif"],
    "age":[12, 20, 19],
    "city":["mardan", "ibd", "mhmnd"]
}
df=pd.DataFrame(data)
df.rename(columns={'name':' STU_Name'}, inplace=True)
df['New_age']=df['age'].astype(float)
df['New_age']=df['age'].apply(lambda x:x*2)
print(df)
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Ayesha", "Bilal", "Hamza", "Zain"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 60000, np.nan, 55000, 65000, 70000, np.nan],
    "Experience": [2, 4, 1, 3, 5, 6, 2]
}

df = pd.DataFrame(data)
df.fillna(0)
df['Salary_fillna']=df['Salary'].fillna(df['Salary'].mean())
print(df)
```

```python
import pandas as pd
import requests
from io import StringIO

url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"

headers = {
    "User-Agent": "Mozilla/5.0"
}

html = requests.get(url, headers=headers).text

tables = pd.read_html(StringIO(html))

print(len(tables))
print(tables[0].head())
```
