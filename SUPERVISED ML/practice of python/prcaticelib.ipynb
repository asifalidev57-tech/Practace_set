```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
df = pd.read_csv("student.csv")

# Feature Engineering
df["Total"] = df["Math"] + df["Physics"] + df["Chemistry"]
df["Percentage"] = df["Total"] / 300 * 100

# Top Student
topper = df.loc[df["Percentage"].idxmax()]

print("Top Student")
print(topper["Name"])
print(topper["Percentage"])

# Summary Statistics
print(df.describe())

# Histogram
plt.figure(figsize=(8,5))
sns.histplot(df["Percentage"], bins=8)
plt.title("Percentage Distribution")
plt.savefig("marks_distribution.png")
plt.show()
plt.close()

# Gender Comparison
plt.figure(figsize=(8,5))
sns.boxplot(x="Gender", y="Percentage", data=df)
plt.title("Percentage by Gender")
plt.savefig("gender_comparison.png")
plt.show()

plt.close()

# Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(
    df[["Math","Physics","Chemistry","Attendance","Study_Hours"]].corr(),
    annot=True
)
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png")
plt.show()
plt.close()
```

```python
import sqlite3

connection = sqlite3.connect('student.db')
cursor = connection.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS student(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL
)
''')

connection.commit()

cursor.execute("""
INSERT INTO student(name)
VALUES('khan')
""")

cursor.execute("""
INSERT INTO student(name)
VALUES('asif')
""")

connection.commit()

cursor.execute("SELECT * FROM student")
rows = cursor.fetchall()

for row in rows:
    print(row)
```
