```python
class teacher:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def student_name(self):
        print(f"The name of the studn is'{self.name}'")    

    def student_age(self):
        print(f"Age of the student is: '{self.age}'") 
class student(teacher):
    def __init__(self,name,age,rollno):
        teacher.__init__(self,name,age)
        self.rollno=rollno

    def show_rolno(self):
        print(f"Rollno of the student is '{self.rollno}'")    

s=student("khan", 19, 9)

s.student_name()
s.student_age()
s.show_rolno()
```

```python
class Animal:
    def sound(self):
        print("every animal has its own style of barking")

class Dog(Animal):
    def sound(self):
        print("Dog bark")

class Cat(Animal):
    def sound(self):
        print("cat meow meow") 



def make_sound(animal):
    animal.sound()  


make_sound(Dog())
make_sound(Cat())


```

```python
class Bank:
    def __init__(self,salary, balance):
        self.__salary=salary
        self.__balance=balance

    def employee_salary(self):
        print(f"The salary of the employee is: {self.__salary}")

    def employee_balance(self):
        print(f"the left balance in his account is {self.__balance}")    

b=Bank(80000, 200000)

b.employee_salary()   
b.employee_balance()         
```
