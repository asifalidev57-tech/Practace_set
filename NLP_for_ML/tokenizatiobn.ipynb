```python
corpus="""
Hello everyone. My name is Asif. I am learning NLP.
"""
from nltk.tokenize import sent_tokenize
sentence=sent_tokenize(corpus)
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import wordpunct_tokenize
for word in sentence:
    print("word tokens: ",word_tokenize(word))
    print("wordpunct tokens ", wordpunct_tokenize(word))
```

```python
from nltk.stem import PorterStemmer
porter=PorterStemmer()
words=['playing',
'played',
'player',
'plays']
for word in words:
    print(porter.stem(word))
```

```python
from nltk.stem import RegexpStemmer
stemer=RegexpStemmer('ing$|s$|e$|able$', min=4)
gpt_sent=['running','runner','runs','ran']
for sent in gpt_sent:
    print(sent, "----->", stemer.stem(sent))
```

```python
from nltk.stem import SnowballStemmer
snowball=SnowballStemmer('english')
```

```python
words=['playing',
'played',
'player',
'plays']
for word in words:
    print(snowball.stem(word))
```
