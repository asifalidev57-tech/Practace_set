```python
corpus="""
The children were running quickly through the beautiful gardens while their parents watched them happily.
Some people were studying different languages, others were writing reports, and a few were discussing the latest technologies.
The scientists had discovered several interesting findings after analyzing the collected data carefully. 
Students are learning, reading, and practicing every day to improve their knowledge and become successful professionals.
The cats were chasing mice, the birds were flying above the trees, 
and the dogs barked loudly because they had seen strangers approaching the house.
"""
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
stemmer=PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(corpus)
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=''.join(words)
```

```python
paragraph="""
Yesterday, the teachers were explaining difficult concepts while the students were taking detailed notes and asking thoughtful questions.
After the lecture, everyone gathered in the library to read books, complete assignments, and prepare for the upcoming examinations.
One researcher was comparing different machine learning algorithms, while another was evaluating their performance using several datasets.
The developers had already fixed the reported bugs, improved the application's security, and optimized its performance before releasing the latest version.
Meanwhile, children were playing outside, women were shopping at the market, men were driving carefully through crowded streets, and elderly people were enjoying the pleasant weather in the nearby park."""
```

```python
sentences=nltk.sent_tokenize(paragraph)
nltk.download('averaged_perceptron_tagger_eng')
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[word for word in words if word not in set(stopwords.words('english'))]
    pos_tag=nltk.pos_tag(words)
    print(pos_tag)
```

```python
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
nltk.ne_chunk(pos_tag).draw()
```
